```python
# example of random oversampling to balance the class distribution
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler, SMOTE
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

# Train dataset

```python
mainfolder = "../data_2019_processed/"
outputFolders = "train/"

#import data
dataset = pd.read_csv(mainfolder+outputFolders+"allData-classification-numeric-normalized.csv")
#del dataset['showActuation']
#del dataset['distanceHome']
```

```python
# Separate the class from the inputs
#X = dataset.iloc[:,:-1]
#y = dataset.iloc[:,-1]
y = dataset['class']
X = dataset.copy()
del X['class']
del X['timestamp_text']
del X['timestamp']
```

```python
print("Train dataset")
# summarize class distribution
print("Before oversampling: ", Counter(y))
```

```python
# define oversampling strategy
#oversample = RandomOverSampler(sampling_strategy='minority')
#oversample = RandomOverSampler(sampling_strategy='auto')
oversample = SMOTE(random_state = 32)
#X_smote_res, y_smote_res = smote.fit_resample(X, y)
```

```python
#y
```

```python

# fit and apply the transform
X_over, y_over = oversample.fit_resample(X, y)
```

```python
# summarize class distribution
print("After undersampling: ", Counter(y_over))
```

```python
dfBalanced = X_over
dfBalanced['class'] = y_over

dfBalanced
```

```python
dfBalanced.to_csv(mainfolder+outputFolders+"allData-classification-numeric-normalized_balanced_oversample_smote.csv", sep=',', encoding='utf-8', index=False)
```

# Test dataset

```python
outputFolders = "test/"

#import data
dataset = pd.read_csv(mainfolder+outputFolders+"allData-classification-numeric-normalized.csv")
#del dataset['showActuation']
#del dataset['distanceHome']
```

```python
# Separate the class from the inputs
#X = dataset.iloc[:,:-1]
#y = dataset.iloc[:,-1]
y = dataset['class']
X = dataset.copy()
del X['class']
```

```python
print("Train dataset")
# summarize class distribution
print("Before oversampling: ", Counter(y))
# define oversampling strategy
oversample = RandomOverSampler(sampling_strategy='auto') #minority
# fit and apply the transform
X_over, y_over = oversample.fit_resample(X, y)
# summarize class distribution
print("After undersampling: ", Counter(y_over))
```

```python
dfBalanced = X_over
dfBalanced['class'] = y_over

dfBalanced
```

```python
dfBalanced.to_csv(mainfolder+outputFolders+"allData-classification-numeric-normalized_balanced_oversample_smote.csv", sep=',', encoding='utf-8', index=False)
```
