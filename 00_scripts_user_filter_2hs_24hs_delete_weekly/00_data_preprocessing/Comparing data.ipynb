```python
# take the list of directories and concat them
def loadDataFromFolders(foldersToLoad,inputFolders,fileType = "_transformed"):
    print(len(foldersToLoad), "datasets")
    for i in range(0,len(foldersToLoad)):
        currentFolder = foldersToLoad[i]
        print(i , "-", currentFolder,inputFolders+"/student_"+currentFolder+""+fileType+".csv")
        #print(trainingDataSet[i])
        if(i == 0):
            temp_data = pd.read_csv(inputFolders+"/student_"+currentFolder+""+fileType+".csv")
        else:
            dataset = pd.read_csv(inputFolders+"/student_"+currentFolder+""+fileType+".csv")
            temp_data = pd.concat([temp_data, dataset])
    # return the dataset        
    return temp_data
```

```python

df = dataAll

import pandas as pd
import numpy as np

# Suponha que você já tenha carregado seus dados em um DataFrame do Pandas chamado df
# Considere que 'df' tem colunas 'activity', 'location', 'day_of_week', etc. para características de entrada,
# e uma coluna 'class' para a classe de saída.

# Calcular a distribuição de Dirichlet para cada característica de entrada em cada classe

# Lista para armazenar os parâmetros alfa para cada característica de entrada e classe
dirichlet_parameters = {}

# Loop sobre as características de entrada
for feature in inputFeatures:
    # Dicionário para armazenar os contadores de ocorrências para cada valor único desta característica em cada classe
    feature_counts = {}

    # Loop sobre as classes de saída
    for output_class in outputClasses:
        # Filtrar o DataFrame para esta classe de saída
        class_df = df[df['class'] == output_class]

        # Contar as ocorrências de cada valor único desta característica nesta classe
        counts = class_df[feature].value_counts().to_dict()

        # Adicionar os contadores a feature_counts
        feature_counts[output_class] = counts

    # Adicionar feature_counts ao dicionário de parâmetros
    dirichlet_parameters[feature] = feature_counts

# Exibir os parâmetros alfa estimados
print("Parâmetros alfa estimados para a distribuição de Dirichlet:")
print(dirichlet_parameters)
```

```python
# Function to calculate the α parameters
def calculate_alpha(dirichlet_parameters):
    alphas = {}
    for feature, counts in dirichlet_parameters.items():
        # Number of classes
        num_classes = len(counts)
        # Number of values for this feature
        num_values = len(set.union(*[set(class_counts.keys()) for class_counts in counts.values()]))
        # Initialize alphas for this feature
        alphas[feature] = np.ones((num_classes, num_values))
        # Iterate over classes
        for i, class_counts in enumerate(counts.values()):
            # Iterate over values
            for j, count in enumerate(class_counts.values()):
                alphas[feature][i, j] += count
    return alphas

# Example usage:
alpha = calculate_alpha(dirichlet_parameters)
print(alpha)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import dirichlet

# Define the α parameters for each class
alpha_class_1 = [2, 2, 2]  # Example α parameters for class 1
alpha_class_2 = [3, 1, 1]  # Example α parameters for class 2

# Sample from the Dirichlet distribution for each class
samples_class_1 = dirichlet.rvs(alpha_class_1, size=1000)
samples_class_2 = dirichlet.rvs(alpha_class_2, size=1000)

# Plot the samples for each class
plt.figure(figsize=(8, 6))
plt.hist(samples_class_1, bins=30, alpha=0.5, label=['Feature 1', 'Feature 2', 'Feature 3'])
plt.hist(samples_class_2, bins=30, alpha=0.5, label=['Feature 1', 'Feature 2', 'Feature 3'])
plt.title('Samples from Dirichlet Distribution for Different Classes')
plt.xlabel('Probability')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
