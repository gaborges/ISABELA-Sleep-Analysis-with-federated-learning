# Data pre-processing

```python
print("Loading data")

allData = pd.DataFrame()

clientData = []

for indexFolder, folder in folders.iterrows():

    # check f the path exists (becouse kaggle removed the + character from file names)
    fileName = folder['folder']
    if(not(exists("../../Dataset_ECUADOR_2019/"+folder['folder']+"/"+fileName+".csv"))):
        fileName = folder['folder'].replace("+","")
    
    # data sensors
    data1 = pd.read_csv("../../Dataset_ECUADOR_2019/"+folder['folder']+"/"+fileName+".csv",
                       delimiter=',', encoding="utf8")
    
    # data from sleep detection
    data2 = pd.read_csv("../../Dataset_ECUADOR_2019/"+folder['folder']+"/Sleep_Detection_"+fileName+".csv",
                       delimiter=',', encoding="utf8")
   
    # merge data
    mergedDS = pd.merge(data1, data2, on='timestamp')
    print(folder['folder']+" "+str(mergedDS.shape[0]))
    # remove unecessary attributes not relevant to sleep detection: foregroundApp,wifi,bluetooth,ble, risk,suburb,municipality,city,town,county
    del mergedDS["foregroundApp"]
    del mergedDS["wifi"]
    del mergedDS["bluetooth"]
    del mergedDS["ble"]
    del mergedDS["risk"]
    del mergedDS["suburb"]
    del mergedDS["municipality"]
    del mergedDS["city"]
    del mergedDS["town"]
    del mergedDS["county"]
    del mergedDS["Unnamed: 0"]
    del mergedDS["activity_y"]
    # rename
    mergedDS.rename(columns = {'activity_x':'activity'}, inplace = True)
    
    # convert merged DF timestamp to datetime
    mergedDS['timestamp_text'] = pd.to_datetime(mergedDS['timestamp'])
    mergedDS['timestamp']  = mergedDS['timestamp_text'].astype(int).div(10**9)
    # time of the day
    #mergedDS['minutes_day'] = mergedDS['timestamp'].hour * 60 + mergedDS['timestamp'].minute
    mergedDS['minutes_day'] = mergedDS.timestamp_text.dt.hour * 60 + mergedDS.timestamp_text.dt.minute
    # initialize times
    mergedDS['sleep_time'] = 0
    mergedDS['wakeup_time'] = 0
    mergedDS['time_sleeping'] = 0
    
    # remove temp information
    #del mergedDS['sleep_time']
    #del mergedDS['wakeup_time']
    #del mergedDS['time_sleeping']
    # replace the place from the class
    
    # remove the outlier (we start the experiment in 1526043855)
    outliers = mergedDS.loc[mergedDS['timestamp'] < 1526043855]
    print("Outliers: ",len(outliers))
    if(len(outliers) > 0):
        print((outliers))
    mergedDS = mergedDS.loc[mergedDS['timestamp'] >= 1526043850]
    # print the guy
    #mergedDS.to_csv(outputFolders+"student_"+folder['folder']+".csv", sep=',', encoding='utf-8', index=False)
    # add in the list
    clientData.append(mergedDS)
    #mergedDS.to_csv("dados2019/student_"+folder['folder']+".csv", sep=',', encoding='utf-8', index=False)
    print("Louded data: ",len(mergedDS))
```

```python
print("labeling data")
for index in range(0,len(clientData)):
#for index in [0,1]:
    currentClientData = clientData[index]
    currentFolder = datasets[index]
    
    dataFLabel = pd.read_csv("../data_2019_processed/form/sleep_form_"+currentFolder+"_treated.csv",
                   delimiter=',', encoding="utf8")
    
    # transform the timestamp on datetime
    dataFLabel['timestamp'] = pd.to_datetime(dataFLabel['timestamp'])
    # sleep_time	wakeup_time
    #dataFLabel['sleep_time'] = pd.to_datetime(dataFLabel['sleep_time'])
    
    #dataFLabel['wakeup_time'] = pd.to_datetime(dataFLabel['wakeup_time'])
    
    # initialize other stuff sleep	arousal	valence	dominance	physical_activity	sociability	study
    currentClientData['sleep_time'] = 0
    currentClientData['wakeup_time'] = 0
    currentClientData['sleep'] = 0
    currentClientData['arousal'] = 0
    currentClientData['valence'] = 0
    currentClientData['dominance'] = 0
    currentClientData['physical_activity'] = 0
    currentClientData['sociability'] = 0
    currentClientData['study'] = 0
    
    # add class
    currentClientData["class"] = 'awake'
    
    last_wakeup_time = 0
    # sleep_time	wakeup_time
    for index, labelSet in dataFLabel.iterrows():
        sleep_time = dataFLabel['sleep_time'][index]
        wakeup_time = dataFLabel['wakeup_time'][index]
        
        #condition = (currentClientData['timestamp'] >= sleep_time) & (currentClientData['timestamp'] <= wakeup_time)
        #currentClientData[condition] = currentClientData[condition].apply(lambda x: addKeyWords(x, list_orange))
        
        # replace value
        for indexT, contentT in currentClientData['timestamp'].items():
            # when attributed, stop the process, unless is the last index
            if(contentT > wakeup_time and index < dataFLabel.index[-1]):
                break;
                
            # attribute the sleep class
            if(sleep_time <= contentT) and (contentT <= wakeup_time):
                currentClientData['class'][indexT] = 'asleep'
                
            # only start after the last wake up
            if(last_wakeup_time > contentT):
                # attribute other stuff
                sleep_time_dt = datetime.fromtimestamp(sleep_time, tz = None)     
                wakeup_time_dt = datetime.fromtimestamp(wakeup_time, tz = None)        
                currentClientData['sleep_time'][indexT] = sleep_time_dt.hour * 60 + sleep_time_dt.minute 
                currentClientData['wakeup_time'][indexT] = wakeup_time_dt.hour * 60 + wakeup_time_dt.minute 
                currentClientData['time_sleeping'][indexT] = wakeup_time - sleep_time 
                # data form the form
                currentClientData['sleep'][indexT]      = dataFLabel['sleep'][index]
                currentClientData['arousal'][indexT]    = dataFLabel['arousal'][index]
                currentClientData['valence'][indexT]    = dataFLabel['valence'][index]
                currentClientData['dominance'][indexT]  = dataFLabel['dominance'][index]
                currentClientData['physical_activity'][indexT] =dataFLabel['physical_activity'][index]
                currentClientData['sociability'][indexT] = dataFLabel['sociability'][index]
                currentClientData['study'][indexT]      = dataFLabel['study'][index]
            
        last_wakeup_time = wakeup_time
            
    print(Counter(currentClientData["class"]))
            
```

# Weekly Division

```python
# identify how many weeks the data has
GlobalMaxValueFromTimestamp = allData["timestamp"].max()
GlobalMinvalueFromTimestamp = allData["timestamp"].min()

print(GlobalMaxValueFromTimestamp,GlobalMinvalueFromTimestamp)

experimentTime = GlobalMaxValueFromTimestamp - GlobalMinvalueFromTimestamp
minuts = experimentTime / (60)
hours = minuts / (60)
days  = hours / 24
weeks = days / 7

print(experimentTime)
print(hours,days,weeks)
```

```python
WeekTime = 60*60*24*7 # min, hour, day, week

print("Week in seconds:",WeekTime)
print()


experimentTime / WeekTime

numberOfWeeks = 4 # each week will have the WeekTime, but the rest not

for i in range(0,numberOfWeeks):
    beginValueWeek = GlobalMinvalueFromTimestamp + (WeekTime*i)
    endValueWeek = GlobalMinvalueFromTimestamp + (WeekTime*(i+1))
    
    if(i == (numberOfWeeks-1)):
        endValueWeek = GlobalMaxValueFromTimestamp
    
    print("week:", i)
    
    print("Begin/end: ",beginValueWeek,endValueWeek)
    amountX = endValueWeek - beginValueWeek
    amountX = amountX / WeekTime
    print("Amount: ",amountX)
    
    print("Creating folder")
    folderName = outputFolders + "w" + str(i)
    isExist = os.path.exists(folderName)
    if not isExist:
        # Create a new directory because it does not exist
        os.makedirs(folderName)
        print("The directory is created!")
    else:
        print("The directory exists!")
    print()
    
    
```
